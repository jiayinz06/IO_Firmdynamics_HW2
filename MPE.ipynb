```julia
using LinearAlgebra
```

```julia
N=40
δ = 1.2
β=1/1.05
α=0.5;
```

```julia
pi = zeros(N,N)
for i in 1:N
    for j in 1:N
        if i<j
            pi[i,j]= 0
        else
            pi[i,j] = δ*(i-j)
        end
    end
end
```

```julia
function findVx(V0, x0)
    V = V0
    x = x0
    
    for i in 1:N
        for j in 1:N
            if j<N && i<N
                W1 = zeros(N)
                W1[i] = V0[i,j]/(1+α*x0[j,i])+ α*x0[j,i]*V0[i,j+1]/(1+α*x0[j,i])
                W1[i+1] = V0[i+1,j]/(1+α*x0[j,i])+ α*x0[j,i]*V0[i+1,j+1]/(1+α*x0[j,i])
                if W1[i+1]-W1[i]>0
                    x[i,j] = max((sqrt(β*α*(W1[i+1]-W1[i]))-1)/α,0)
                else
                    x[i,j] =0
                end
                V[i,j] = pi[i,j]-x[i,j]+β*(W1[i]/(1+α*x[i,j])+α*x[i,j]*W1[i+1]/(1+α*x[i,j]))
            elseif j==N && i<N
                W1 = zeros(N)
                W1[i] = V0[i,N]
                W1[i+1] = V0[i+1,N]
                if W1[i+1]-W1[i]>0
                    x[i,j] = max((sqrt(β*α*(W1[i+1]-W1[i]))-1)/α,0)
                else
                    x[i,j] =0
                end
                V[i,j] = pi[i,j]-x[i,j]+β*(W1[i]/(1+α*x[i,j])+α*x[i,j]*W1[i+1]/(1+α*x[i,j]))
            elseif j<N && i==N
                W1 = zeros(N)
                W1[i] = V0[i,j]/(1+α*x0[j,i])+ α*x0[j,i]*V0[i,j+1]/(1+α*x0[j,i])
                x[i,j] = 0
                V[i,j] = pi[N,j]-x[i,j]+β*W1[N]
            else 
                W1 = zeros(N)
                W1[i] = V0[i,N]
                x[i,j] = 0
                V[i,j] = pi[N,j]-x[i,j]+β*W1[N]
            end
        end
    end  
    return V,x   
end
```

```julia
function findV(V0,x0,max_it = 1e4)
    i = 1
    dist = 1e3
    V1 = V0
    x1 = x0
    while (i < max_it) 
        (V,x)=findVx(V1, x1);
        dist = norm(V-V1)
        V1 = V
        x1 = x
        i += 1
    end
    return V1, x1
end
```

```julia
V0 = rand(N,N)
x0 = rand(N,N)
(V1, x1) = findV(V0,x0);
        
```
